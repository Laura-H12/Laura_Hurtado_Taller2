```python
plt.figure(figsize=(50,20))
plt.subplot(121)
nx.draw(g, with_labels=True,
       pos=nx.fruchterman_reingold_layout(g),
       node_size=60, 
        node_color="skyblue", 
        node_shape="s", 
        alpha=0.5, 
        linewidths=40,
        font_weight='bold')

plt.show()
```
